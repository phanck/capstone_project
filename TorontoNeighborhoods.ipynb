Import Libraries

```python
import pandas as pd
```

### Create Notebook and Scrape wikipedia
sections 1 and 2: create notebook and scrape the wikipedia page

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
dfs = pd.read_html(url, header=0) # set header with row #0
# M_postal_codes_canada = MPCC
MPCC = dfs[0]
```

### Massage the data
section 3: create dataframe

```python
MPCC = MPCC[MPCC.Borough != 'Not assigned']
MPCC.head(10)
```

### Add Latitude and Longitude
section 4: use pgeocode to get latitude and longitude

```python
import pgeocode
```

```python
nomi = pgeocode.Nominatim('ca')
pc = nomi.query_postal_code("M3A")
#print("lat {} long {}".format(pc.latitude,pc.longitude))
MPCC1 = MPCC.copy()
```

```python
MPCC1.head(10)
```

```python
MPCC1['Latitude'] = MPCC1['Postal Code'].apply(lambda x : nomi.query_postal_code(x)['latitude'])
MPCC1['Longitude'] = MPCC1['Postal Code'].apply(lambda x : nomi.query_postal_code(x)['longitude'])
```
